```python
import numpy as  np
import scipy
import matplotlib.pyplot as plt
```

```python
def pol(x,w,w0,Q,f,gamma):
    return x**6*(9*gamma**2/(16*w0**2))+x**4*(3*gamma*(1-w**2/w0**2)/2)+x**2*((w/(Q*w0))**2+(1**2-w**2/w0**2)**2) -f**2/w0**2
```

```python
w = np.linspace(5e6,30e6,500)
Q = 200
w0 = 14e6
f = 1.2e-11/1e-16/1e-9
gamma = 5e30*1e-18

x0=np.array([1,1,1,1,1,1])

amplitude = []
for w1 in w:
    amplitude.append(scipy.optimize.fsolve(pol,x0,args=(w1,w0,Q,f,gamma),maxfev=10000))
```

```python
plt.plot(w,amplitude)
plt.show()
```
